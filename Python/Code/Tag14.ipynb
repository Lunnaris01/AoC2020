```python
import numpy as np
import pandas as pd
data = np.array(pd.read_csv("../../res/input_14.txt",header=None))
```

```python
print(data)
```

```python
mem = dict()
```

```python
def bin_list_from_decimal(dec_val: str) -> np.array:
    return list(f"{int(bin(int(dec_val))[2:]):036d}")

def apply_mask_to_bin_array(bin_arr: np.array,mask :np.array) -> np.array:
    for i in range(len(mask)):
        if mask[i] != 'X':
            bin_arr[i]=mask[i]
    return bin_arr
```

```python
currentmask = list(data[0][0].split(" = ")[1])
for line in data[1:]:
    cmd, val = line[0].split(" = ")
    if(cmd[0:3] == 'mem'):
        adress = cmd[4:-1]
        mem[adress] = apply_mask_to_bin_array(bin_list_from_decimal(val),currentmask)
    if(cmd[0:4] == 'mask'):
        currentmask = val
```

```python
memsum = 0
for item in mem.values():
    itemval = 0
    for i in range(len(item)):
        itemval+= 2**(i)*int(item[::-1][i])
    memsum+=itemval
```
