```python
def countencounters(data,right,down):
    xpos=0
    ypos=0
    counter = 0
    while(ypos<(data.shape[0]-1)):
        xpos = (xpos+right)%(len(data[0][0]))
        ypos+=down
        if(data[ypos][0][xpos] == '#'):
            counter+=1
    return counter
```

```python
print(countencounters(data,3,1))
```
