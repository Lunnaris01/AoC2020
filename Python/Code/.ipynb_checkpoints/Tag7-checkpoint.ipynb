```python
from collections import defaultdict
import numpy as np
import pandas as pd
data = np.array(pd.read_csv("../../res/input_7.txt",header=None).fillna("NAN"))
for text in data:
    print(text)
```

```python
bags  = defaultdict(list)
numbers = ['1','2','3','4','5']
for lines in data:
    outer,inner = lines[0].split('contain ');
    outer = outer[0:-6]
    inner, _ = inner.split(' bag')
    if (inner[0] in numbers):
        bags[inner[2:]].append(outer)
    for line in lines[1:]:
        if(line != 'NAN'):
            inline,_ = line.split(' bag')
            bags[inline[3:]].append(outer)
print(bags['shiny gold'])
```

```python
stack = list(bags['shiny gold'])
baglist = []
while(len(stack) != 0):
    current = stack.pop()
    for bag in bags[current]:
        stack.append(bag)
    baglist.append(current)
```

```python
print(len((set(baglist))))
```

```python
baglist  = defaultdict(list)
numbers = ['1','2','3','4','5']
for lines in data:
    outer,inner = lines[0].split('contain ');
    outer = outer[0:-6]
    inner, _ = inner.split(' bag')
    if (inner[0] in numbers):
        baglist[outer].append(inner)
    for line in lines[1:]:
        if(line != 'NAN'):
            inline,_ = line.split(' bag')
            baglist[outer].append(inline[1:])
print(baglist['shiny gold'])
```

```python
def amountofbags(baglist,bag):
    cost = 1
    if(len(baglist[bag])>0):
        for subbag in baglist[bag]:
            cost = cost + amountofbags(baglist,subbag[2:])*int(subbag[0])
    return cost
        
```

```python
print(amountofbags(baglist,'shiny gold')-1) # -1 because we only look at the bags inside our bag not the bag itself!
```
