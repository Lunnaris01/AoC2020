```python
import numpy as np
import pandas as pd
import re
data = np.array(pd.read_csv("../../res/input_6.txt",header=None,skip_blank_lines = False,error_bad_lines=False).fillna("NEXTGROUP"))
for line in data:
    print(line[0])
```

```python
# Teil 1
countinglist=[]
current=[]
for text in data:
    if(text[0] == "NEXTGROUP"):
        countinglist.append(len(current))
        print(sorted(current))
        current.clear()
    else:
        for letter in text[0]:
            if letter not in current:
                current.append(letter)
countinglist.append(len(current))
```

```python
# Teil 2
countinglist=[]
temp_0 = []
temp_1 = []
nextgroup = 1
for text in data:
    if(text[0] == "NEXTGROUP"):
        countinglist.append(len(temp_0))
        print(sorted(temp_0))
        temp_0.clear()
        nextgroup = 1
    else:
        temp_1.clear()
        print(text[0])
        for letter in text[0]:
            temp_1.append(letter)
        if(nextgroup == 1):
            temp_0 = list(temp_1)
            nextgroup = 0
        temp_2 = list(temp_0)
        for letter in temp_0:
            if letter not in temp_1:
                temp_2.remove(letter)
        temp_0 = list(temp_2)
print(temp_0)
countinglist.append(len(temp_0))
```

```python
print(np.sum(countinglist))
```

```python
def berechne(inputs):
    alphabet = []
    for letter in range(97,123):
        alphabet.append(chr(letter))
    dump = []
    for letter in range(97,123):
        dump.append(chr(letter))

    for word in inputs:
        for letter in alphabet:
            if(letter not in word):
                if(letter in dump):
                    dump.remove(letter)
    print(dump)
    return(len(dump))
```

```python
countlist = []
sublist = []
for text in data:
    if text[0] == "NEXTGROUP":
        countlist.append(berechne(sublist))
        sublist.clear()
    else:
        sublist.append(text[0])
countlist.append(berechne(sublist))

print(np.sum(countlist))
```
