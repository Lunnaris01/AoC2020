```python
import numpy as np
import pandas as pd
```

```python
data = np.array(pd.read_csv("../res/input_2.txt",header=None))
```

```python
print(data)
```

```python
def preprocess(line):
    policy,password = line.split(':')
    constraint,letter = policy.split()
    minval,maxval = constraint.split('-')
    minval = int(minval)
    maxval = int(maxval)
    return(minval,maxval,letter,password)

def checkpassword(line):
    minval,maxval,letter,password=preprocess(line)
    counter = 0
    for letterinpass in password:
        if letter == letterinpass:
            counter+=1
    
    if (counter >= minval and counter <= maxval):
        return 1
    return 0

actually_counter = 0;
for i in data:
    actually_counter+=checkpassword(i[0])
print(actually_counter)
```

```python
def checkpassword_2(line):
    firstval,secondval,letter,password=preprocess(line)
    if(password[firstval-1] == letter):
        if(password[secondval-1] ==letter):
            return 0
        else:
            return 1
    else:
        if(password[secondval-1] ==letter):
            return 1
    return 0

actually_counter = 0;
for i in data:
    actually_counter+=checkpassword_2(i[0])
print(actually_counter)
```
